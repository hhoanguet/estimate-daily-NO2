## FEATURE ENGINEERING

### Handle missing value and outliers in NO2 observations

Các hàng không có quan sát NO2 sẽ được loại bỏ.

```python
dataset = dataset.dropna(subset=['NO2'])
```

Loại bỏ 2 giá trị nhiễu là 651519 và 37321 ở cột NO2 observation như đã trinh bày trong báo cáo.

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(dataset.NO2.values))
threshold = 3
print(np.where(z>3))
print(np.where(z<-3))
print("Outlier values: ", dataset.NO2.iloc[716], dataset.NO2.iloc[857])
```

```python
dataset.NO2.replace(651519.076000,np.nan, inplace=True)
dataset.NO2.replace(37321.263950,np.nan, inplace=True)
```

### Loại bỏ giá trị -9999 do không đọc được dữ liệu ở cột pd

```python
dataset.pd.describe()
```

```python
dataset.pd.replace(-9999,np.nan, inplace=True)
```

```python
dataset.pd.describe()
```

### Handle missing values(numerical features)

Fill missing values bằng **giá trị mean** của theo từng cột.

```python
from sklearn.impute import SimpleImputer
num_cols = [col for col in dataset.columns if dataset[col].dtype != 'object']
#Imputation
my_imputer = SimpleImputer()
imputed_data = pd.DataFrame(my_imputer.fit_transform(dataset[num_cols]))
# Imputation removed column names; put them back
imputed_data.columns = ['time', 'lat', 'lon',	'NO2',	'hpbl',	'press',	'rh',
                        'tmp',	'wspd',	'dpt',	'ndvi',	'omi_no2',	'pd',	'rd']
```

```python
imputed_data
```

### Prepare data for fitting the model

```python
# Không dùng cột name như trong báo cáo chúng em đã đề cập
y = imputed_data['NO2'].copy()
X = imputed_data.drop(['NO2'] , axis = 1)
```

### Integrate kết quả cluster của K-means vào data

```python
from sklearn.cluster import KMeans

# Dùng Elbow method để tìm ra số cluster tối ưu, tính WSS (within-cluster sum of squared)
wss = []
K = range(1,9)
for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++').fit(X)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(1, figsize=(16,6))
plt.plot(wss)
plt.xticks(K)
```

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
intergrated_kmean_X = X.copy()
intergrated_kmean_X['cluster'] = kmeans.labels_
```

```python
intergrated_kmean_X.head()
```

```python
intergrated_kmean_X.describe()
```

## Random Forest

```python
from sklearn.model_selection import ShuffleSplit
cv = ShuffleSplit(n_splits=5, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
random_forest_model = RandomForestRegressor(n_estimators=200, random_state=0)
```

```python
rmse_scores = -1 * cross_val_score(random_forest_model, X, y, cv=cv, scoring = 'neg_root_mean_squared_error')
print("RMSE scores:\n", rmse_scores)
print("Average RMSE score:\n", rmse_scores.mean())
```

```python
r2_scores = cross_val_score(random_forest_model, X, y, cv=cv, scoring = 'r2')
print("r2 scores:\n", r2_scores)
print("Average R2 score:\n", r2_scores.mean())
```

# SAVE A TEST.TIF FILE ON CURRENT DIRECTORY

```python
import gdal, ogr, os, osr
import numpy as np
import pandas as pd


def array2raster(newRasterfn,rasterOrigin,pixelWidth,pixelHeight,array):

    cols = array.shape[1]
    rows = array.shape[0]
    originX = rasterOrigin[0]
    originY = rasterOrigin[1]

    driver = gdal.GetDriverByName('GTiff')
    outRaster = driver.Create(newRasterfn, cols, rows, 1, gdal.GDT_Byte)
    outRaster.SetGeoTransform((originX, pixelWidth, 0, originY, 0, pixelHeight))
    outband = outRaster.GetRasterBand(1)
    outband.WriteArray(array)
    outRasterSRS = osr.SpatialReference()
    outRasterSRS.ImportFromEPSG(4326)
    outRaster.SetProjection(outRasterSRS.ExportToWkt())
    outband.FlushCache()

rasterOrigin = (100.1, 25.6)
pixelWidth = 0.025
pixelHeight = 0.025
newRasterfn = 'test.tif'

reversed_arr = no2[::-1]
array2raster(newRasterfn,rasterOrigin,pixelWidth,pixelHeight,reversed_arr)
```
